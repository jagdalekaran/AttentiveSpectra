```python
import os
import numpy as np
import h5py
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.layers import Layer, Multiply, Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, LSTM, Add, Activation, Reshape, Permute
from tensorflow.keras.models import Model
from tensorflow.keras.utils import Sequence
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint, LearningRateScheduler
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import AdamW
import random

# Define the SpectralAttentionLayer
class SpectralAttentionLayer(Layer):
    def __init__(self, num_bands, **kwargs):
        super(SpectralAttentionLayer, self).__init__(**kwargs)
        self.num_bands = num_bands

    def build(self, input_shape):
        self.dense1 = tf.keras.layers.Dense(self.num_bands // 8, activation='relu', kernel_regularizer=l2(0.001))
        self.dense2 = tf.keras.layers.Dense(self.num_bands, activation='sigmoid', kernel_regularizer=l2(0.001))

    def call(self, inputs):
        avg_pool = tf.reduce_mean(inputs, axis=[1, 2], keepdims=False)
        attention = self.dense1(avg_pool)
        attention = self.dense2(attention)
        attention = tf.reshape(attention, (-1, 1, 1, self.num_bands))
        self.attention_weights = attention
        output = Multiply()([inputs, attention])
        return output

    def compute_output_shape(self, input_shape):
        return input_shape

    def get_attention_weights(self):
        return self.attention_weights

# Custom data augmentation
def augment_patch(patch):
    if random.random() > 0.5:
        patch = np.flip(patch, axis=0)  # Horizontal flip
    if random.random() > 0.5:
        patch = np.flip(patch, axis=1)  # Vertical flip
    if random.random() > 0.5:
        angle = random.uniform(-20, 20)  # Random rotation between -20 and 20 degrees
        patch = tf.keras.preprocessing.image.random_rotation(patch, angle, row_axis=0, col_axis=1, channel_axis=2)
    if random.random() > 0.5:
        noise = np.random.normal(0, 0.01, patch.shape)  # Adding Gaussian noise
        patch = patch + noise
    return np.clip(patch, 0, 1)  

# Data generator with custom augmentation and oversampling
class AugmentedOversampledDataGenerator(Sequence):
    def __init__(self, filepaths, labels, batch_size=32, patch_size=(180, 180), shuffle=True, augment=False, oversample_minority=True):
        self.filepaths = filepaths
        self.labels = labels
        self.batch_size = batch_size
        self.patch_size = patch_size
        self.shuffle = shuffle
        self.augment = augment
        self.oversample_minority = oversample_minority

        if oversample_minority:
            self._oversample_minority_class()

        self.on_epoch_end()

    def _oversample_minority_class(self):
        # Oversample minority class (label == 1)
        minority_indices = [i for i, label in enumerate(self.labels) if label == 0]
        majority_indices = [i for i, label in enumerate(self.labels) if label == 1]

        oversample_count = len(majority_indices) - len(minority_indices)
        if oversample_count > 0:
            oversample_indices = np.random.choice(minority_indices, oversample_count, replace=True)
            self.filepaths += [self.filepaths[i] for i in oversample_indices]
            self.labels += [self.labels[i] for i in oversample_indices]

    def __len__(self):
        return int(np.floor(len(self.filepaths) / self.batch_size))

    def __getitem__(self, index):
        batch_filepaths = self.filepaths[index * self.batch_size:(index + 1) * self.batch_size]
        batch_labels = self.labels[index * self.batch_size:(index + 1) * self.batch_size]
        X, y = self.__data_generation(batch_filepaths, batch_labels)
        return X, y

    def on_epoch_end(self):
        if self.shuffle:
            indices = np.arange(len(self.filepaths))
            np.random.shuffle(indices)
            self.filepaths = [self.filepaths[i] for i in indices]
            self.labels = [self.labels[i] for i in indices]

    def __data_generation(self, batch_filepaths, batch_labels):
        X = []
        y = []

        for i, filepath in enumerate(batch_filepaths):
            with h5py.File(filepath, 'r') as h5file:
                intensities = h5file['intensities'][:]
                patches = self.extract_patches(intensities)
                for patch in patches:
                    # Apply augmentation only to the minority class 
                    if self.augment and batch_labels[i] == 0:
                        patch = augment_patch(patch)
                    X.append(patch)
                    y.append(batch_labels[i])

        X = np.array(X)
        y = np.array(y)
        return X, y

    def extract_patches(self, hypercube):
        patches = []
        h, w, num_bands = hypercube.shape
        for i in range(0, h - self.patch_size[0] + 1, self.patch_size[0]):
            for j in range(0, w - self.patch_size[1] + 1, self.patch_size[1]):
                patch = hypercube[i:i+self.patch_size[0], j:j+self.patch_size[1], :]
                if patch.shape[0] == self.patch_size[0] and patch.shape[1] == self.patch_size[1]:
                    patches.append(patch)
        return patches

# Learning rate scheduler
def scheduler(epoch, lr):
    decay_rate = 0.1
    decay_step = 10
    if epoch % decay_step == 0 and epoch:
        return float(lr * decay_rate)
    return float(lr)

# Build the CNN-Attention-RNN hybrid model
def build_cnn_attention_rnn(input_shape, num_classes):
    inputs = Input(shape=input_shape)
    
    # Spectral attention
    spectral_attention = SpectralAttentionLayer(input_shape[2])(inputs)

    # CNN layers
    x = Conv2D(32, (3, 3), padding='same', kernel_regularizer=l2(0.001))(spectral_attention)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)

    x = Conv2D(64, (3, 3), padding='same', kernel_regularizer=l2(0.001))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)

    # Reshape for the RNN
    x = Permute((2, 1, 3))(x)  
    x = Reshape((-1, 64))(x)  

    # RNN for spectral sequence
    x = LSTM(64, return_sequences=False, dropout=0.5)(x)
    
    outputs = Dense(num_classes, activation='softmax')(x)
    
    model = Model(inputs, outputs)
    
    model.compile(optimizer=AdamW(learning_rate=0.001), 
                  loss='sparse_categorical_crossentropy', 
                  metrics=['accuracy'])
    
    return model

# Load filepaths and labels
def load_filepaths_and_labels(cancer_dir, normal_dir):
    cancer_filepaths = [os.path.join(cancer_dir, f) for f in os.listdir(cancer_dir) if f.endswith('.h5')]
    normal_filepaths = [os.path.join(normal_dir, f) for f in os.listdir(normal_dir) if f.endswith('.h5')]
    filepaths = cancer_filepaths + normal_filepaths
    labels = [1] * len(cancer_filepaths) + [0] * len(normal_filepaths)
    return filepaths, labels

# Function to plot training history
def plot_training_history(history):
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Accuracy vs. Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Loss vs. Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Function to plot attention weights against wavenumbers
def plot_attention_weights_vs_wavenumbers(model, data_generator, num_samples=10):
    attention_layer = None
    for layer in model.layers:
        if isinstance(layer, SpectralAttentionLayer):
            attention_layer = layer
            break

    if attention_layer is None:
        print("SpectralAttentionLayer not found in the model.")
        return

    @tf.function
    def get_attention_weights(input_data):
        model(input_data)
        return attention_layer.get_attention_weights()

    available_samples = len(data_generator)
    num_samples = min(num_samples, available_samples)
    sample_indices = random.sample(range(available_samples), num_samples)
    
    wavenumbers = np.linspace(898.659861087799, 3800.9840905666347, attention_layer.num_bands)

    plt.figure(figsize=(12, 8))
    
    for i, idx in enumerate(sample_indices):
        X, _ = data_generator[idx]
        attention_weights = get_attention_weights(X)
        attention_weights = np.mean(attention_weights.numpy(), axis=(0, 1, 2))

        plt.subplot(num_samples, 1, i + 1)
        plt.plot(wavenumbers, attention_weights, label=f'Sample {i + 1}')
        plt.xlabel('Wavenumber (cm^-1)')
        plt.ylabel('Attention Weight')
        plt.legend()

    plt.tight_layout()
    plt.show()

def main():
    # Paths to directories containing hypercubes
    cancer_cubes_dir = "C:/Users/sc23krj/Desktop/Karan MSC Thesis/Training Data/Cancerous Cores"
    normal_cubes_dir = "C:/Users/sc23krj/Desktop/Karan MSC Thesis/Training Data/Normal Cores"
    
    # Load file paths and labels
    filepaths, labels = load_filepaths_and_labels(cancer_cubes_dir, normal_cubes_dir)

    # Split into train and test
    X_train, X_test, y_train, y_test = train_test_split(filepaths, labels, test_size=0.2, random_state=42)

    # Calculate class weights to handle imbalance
    class_weight = {0: len(labels) / labels.count(0), 1: len(labels) / labels.count(1)}

    # Define input shape
    input_shape = (180, 180, 1506)
    num_classes = 2  # Binary classification

    # Create data generators
    train_generator = AugmentedOversampledDataGenerator(X_train, y_train, batch_size=8, augment=True, oversample_minority=True)
    test_generator = AugmentedOversampledDataGenerator(X_test, y_test, batch_size=8, augment=False, oversample_minority=False)

    # Build the model
    model = build_cnn_attention_rnn(input_shape, num_classes)

    # Print model summary
    model.summary()

    # Define callbacks
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=0.00001)
    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    model_checkpoint = ModelCheckpoint('best_spectral_attention_cnn.keras', monitor='val_loss', save_best_only=True)
    lr_scheduler = LearningRateScheduler(scheduler)

    # Train the model using the generator
    history = model.fit(train_generator, epochs=50, validation_data=test_generator,
                        class_weight=class_weight, callbacks=[reduce_lr, early_stopping, model_checkpoint, lr_scheduler])

    # Plot the training history
    plot_training_history(history)

    # Save the trained model
    model.save("spectral_attention_cnn.keras")

    # Evaluate model on test data
    loss, accuracy = model.evaluate(test_generator)
    print(f"Test Loss: {loss}")
    print(f"Test Accuracy: {accuracy}")

    # Predictions on test data
    y_pred = []
    y_true = []
    for i in range(len(test_generator)):
        X, y = test_generator[i]
        predictions = model.predict(X)
        predicted_classes = np.argmax(predictions, axis=1)
        y_pred.extend(predicted_classes)
        y_true.extend(y)

    # Convert to numpy arrays
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    # Additional evaluation metrics
    print("Confusion Matrix:")
    print(confusion_matrix(y_true, y_pred))

    print("\nClassification Report:")
    print(classification_report(y_true, y_pred))

    # Plot the attention weights vs wavenumbers for 10 random samples
    plot_attention_weights_vs_wavenumbers(model, test_generator, num_samples=10)

if __name__ == "__main__":
    main()
```
